```python
gender_counts = df['Q3'].value_counts()
```

```python
gender_counts_dict = gender_counts.to_dict()
```

```python
labels=[]
counts=[]
```

```python
for key , value in gender_counts_dict.items():
    labels.append(key)
    counts.append(value)
```

```python
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(counts)))
fig , ax = plt.subplots()
ax.pie(counts[:2],labels=labels[:2],autopct='%1.1f%%',radius=5,center=(2,2),wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True,colors=colors)
plt.axis('off')
plt.show()
```
